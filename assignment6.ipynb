```python
# 1. Import necessary libraries
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
```

## Prepare data

```python

# 2. Prepare the data
X = 2 * np.random.rand(10000, 5)
y = 3 + np.dot(X, [4.5, 3.2, -1.5, 2.7, 0.5]) + np.random.randn(10000)
```

```python

# Standardization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Build the regression model

```python

# 3. Build the regression model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(5,))
])
```

```python

# 4. Compile the model
model.compile(optimizer='adam', loss='mse')
```

```python

# 5. Train the model
history = model.fit(X_scaled, y, epochs=100, batch_size=32, validation_split=0.2)
```

```python

# 6. Evaluate the model
mse_score = model.evaluate(X_scaled, y)
print("Mean Squared Error (MSE):", mse_score)
```
